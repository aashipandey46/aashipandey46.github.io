**Color representation of Images using Histogramming methods**

```python
fig, ax = plt.subplots(4,1,figsize=(15,15)) # 3 color channels and IMAGE1 plot

ax[0].hist(data1[:,:,0].flatten(), bins=256, color='red') # all columns, all rows, 0th or first channel = Red channel
ax[1].hist(data1[:,:,1].flatten(), bins=256, color='green') # Green Channel
ax[2].hist(data1[:,:,2].flatten(), bins=256, color='blue') # Blue Channel

ax[0].set_xlabel('R channel bins')
ax[1].set_xlabel('G channel bins')
ax[2].set_xlabel('B channel bins')
ax[0].set_ylabel('Number of pixels')
ax[1].set_ylabel('Number of pixels')
ax[2].set_ylabel('Number of pixels')

# finally, include my image:
ax[3].imshow(data1)

plt.show()
```

This is original flower image found on internet. Thus, it has way more colors and more overlapping of bands. This shows that it is quite different from its handmade version(image2).

```python
fig, ax = plt.subplots(4,1,figsize=(15,15)) # 3 color channels and IMAGE2 plot

ax[0].hist(data2[:,:,0].flatten(), bins=256, color='red') # all columns, all rows, 0th or first channel = Red channel
ax[1].hist(data2[:,:,1].flatten(), bins=256, color='green') # Green Channel
ax[2].hist(data2[:,:,2].flatten(), bins=256, color='blue') # Blue Channel

ax[0].set_xlabel('R channel bins')
ax[1].set_xlabel('G channel bins')
ax[2].set_xlabel('B channel bins')
ax[0].set_ylabel('Number of pixels')
ax[1].set_ylabel('Number of pixels')
ax[2].set_ylabel('Number of pixels')

# finally, include my image:
ax[3].imshow(data2)

plt.show()
```

This image is handmade version of image1 above. Thus it has less color overlap and more distinct colors. But it has bit similar distinct color bands as above image (image1).

```python
fig, ax = plt.subplots(4,1,figsize=(15,15)) # 3 color channels and IMAGE1 plot

ax[0].hist(data3[:,:,0].flatten(), bins=256, color='red') # all columns, all rows, 0th or first channel = Red channel
ax[1].hist(data3[:,:,1].flatten(), bins=256, color='green') # Green Channel
ax[2].hist(data3[:,:,2].flatten(), bins=256, color='blue') # Blue Channel

ax[0].set_xlabel('R channel bins')
ax[1].set_xlabel('G channel bins')
ax[2].set_xlabel('B channel bins')
ax[0].set_ylabel('Number of pixels')
ax[1].set_ylabel('Number of pixels')
ax[2].set_ylabel('Number of pixels')

# finally, include my image:
ax[3].imshow(data3)

plt.show()
```

The last image is most different and more overlapping indicating that it has maximum number of shades of color. Also, we have seen that shape of this image is dissimalar to other two images.
